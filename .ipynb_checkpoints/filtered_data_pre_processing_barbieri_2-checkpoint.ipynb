```python
import emoji 
import json
import datetime
import re
from collections import Counter
import timeit
import os
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
#import ijson

file_path = "data_tweets_with_emoji_json.txt"
```

```python
### l√§d Datei
def get_tweets(filename):
    for line in open(filename, encoding='utf-8'):
        yield json.loads(line)
        # yield print(twitter.Status.NewFromJsonDict(json.loads(line)))


def get_full_tweets(filename):
    for line in open(filename, encoding='utf-8'):
        yield json.loads(line)

# dekstop
#data = get_full_tweets("C:\\MSc\\data_tweets_with_emoji_json.txt")

# mobile
data = get_full_tweets("C:\\Users\\Wilm\\Documents\\GitHub\\MSc\\Barbieri Korpus\\data_tweets_with_emoji_json.txt")
```

```python
df = pd.DataFrame(data, columns=['id_str','text','in_reply_to_user_id_str'])
df.info()
```

```python
def extract_chars(my_str):     
    return ''.join([c for c in my_str if not(c in emoji.UNICODE_EMOJI)])

def anonymize_username(my_str):
    anonymized_text = ['@username' if x.startswith('@') else x for x in list(my_str.split())]
    return ' '.join(anonymized_text)

def anonymize_hashtag(my_str): 
    return ' '.join(['#hashtag' if x.startswith('#') else x for x in list(my_str.split())])

def delete_links(my_str): 
    return ' '.join(['' if x.startswith('https') else x for x in list(my_str.split())])
# to do: hashtags und links entfernen

df_clean_text = df['text'].apply(lambda x: delete_links(anonymize_hashtag(anonymize_username(extract_chars(x)))))
```

```python
def extract_emojis(my_str):
    return ''.join(c for c in my_str if c in emoji.UNICODE_EMOJI)

df_extr_emojis = df['text'].apply(lambda x: extract_emojis(x))
```

```python
# unhide the strings in jupyter notebook
pd.options.display.max_rows
pd.set_option('display.max_colwidth', -1)

# assign all new functions to df

new_df = df.assign(clean_text = df_clean_text, extr_emojis = df_extr_emojis) # , langid = df_langid
new_df.head()
```

```python
# rows only with emoji

# len(df_emoji_rows.index) = 194625

df_emoji_rows = new_df[new_df.extr_emojis != '']
df_emoji_rows.head()
```

```python
# POS Tagger "TweeboParser" 

# pre processing for POS Tagger
# df_emoji_rows.text.to_csv(r'barbieri_tweets_text.txt', header=None, index=None, mode='a', sep=',')

# creating dataframe
df_pos = pd.read_csv("C:\\Users\\Wilm\\Documents\\GitHub\\MSc\\ark-tweet-nlp-0.3.2\\barbieri_POS_tagged_tweets.txt", sep='\t')
```

```python
df_pos.columns = ['raw_text', 'POS', 'likelihood', 'tweet_text_2']
df_pos = df_pos.drop(columns = 'tweet_text_2')


df_pos_emo_end = df_pos[df_pos['POS'].str.endswith('E',na = False)]
# counting rows endswith emoji : len(df_pos_emo_end.index) == 36428
df_pos_emo_end.head()
```

```python
# finding the last word category befor emoji 

def last_word_category(my_string):
    s = list(my_string.replace(" ",""))
    categories = []
    index = 0
    
    while index != len(s):
    #print(s[index])
        if s[index] == 'E':
            category = s[index-1]
            categories.append(category)
        index += 1
    return ' '.join(categories)

df_last_word_category = df_pos_emo_end['POS'].apply(lambda x: last_word_category(x))
```

```python
df_pos_emo_end = df_pos_emo_end.assign(last_word_category = df_last_word_category)

df_pos_emo_end.head()
```

```python
# last word category to emoji

df_pos_emo_end = df_pos_emo_end[df_pos_emo_end.last_word_category.str.len() == 1]
df_pos_emo_end.last_word_category.value_counts().head(20)
```
